```python
import sklearn # IMPORTANDO A BIBLIOTECA SCIKIT-LEARN
from sklearn.datasets import load_iris # TRAZENDO DO SCIKIT-LEARN O DATASET IRIS
from sklearn.model_selection import train_test_split # TRAZENDO DO SCIKIT-LEARN UMA FUNÇÃO PARA DIVIDIR OS DADOS EM CONJUNTOS (CONJUNTO DE TREINAMENTO E CONJUNTO DE TESTES DO MODELO)
from sklearn.naive_bayes import GaussianNB # TRAZENDO DO SCIKIT-LEARN O GAUSSIAN NB, QUE TRAZ ATÉ NÓS UM ALGORITMO SIMPLES DE CLASSIFICAÇÃO BINÁRIA (NAIVE BAYES - NB)
from sklearn.metrics import accuracy_score # TRAZENDO DO SCIKIT-LEARN O ACCURACY_SCORE PARA MEDIR A PRECISÃO DO CLASSIFICADOR DE MACHINE LEARNING
```

```python
iris = load_iris() # CRIAÇÃO DO ENTÃO DICIONARIO (CARREGANDO O DATASET)
```

```python
# CRIANDO NOVAS VARIAVEIS PARA CADA CONJUNTO IMPORTANTE DE INFORMAÇÕES E ATRIBUINDO OS DADOS
label_names = iris['target_names'] # Classes de nomes (setosa, versicolor e virginica)
labels = iris['target'] # Alvos para "análise"
feature_names = iris['feature_names'] # Nome do recurso (ou característica) de avaliação / classificação
features = iris['data'] # Dados relacionados aos alvos de acordo com o cumprimento da sepala em centímetros, largura da sépala em centímetros, cumprimento da pétala em centímetros  e largura da pétala em centímetros
```

```python
# TESTES DE VISUALIZAÇÃO DE DADOS
print(label_names)
print(labels)
print(feature_names)
print(features)
```

```python
# ORGANIZANDO OS DADOS EM CONJUNTOS
train, test, train_labels, test_labels = train_test_split(features,
                                                            labels,
                                                            test_size=0.50,
                                                            random_state=42)

# CONSTRUINDO E AVALIANDO O MODELO

# INICIALIANDO O IDENTIFICADOR PARA CONSTRUIR E AVALIAR O MODELO
gnb = GaussianNB() # Inicializando o modelo com a função GaussianNB() depois treine o modelo, ajustando-o aos dados usando gnb.fit()
```

```python
# TREINANDO O CLASSIFICADOR
model = gnb.fit(train, train_labels) # Treinando o modelo e ajustando ele aos dados usando a instanciação gnb.fit()

# FAZENDO AS PREVISÕES
preds = gnb.predict(test) # COM O ENTÃO MODELO TREINADO, PODEMOS FAZER PREVISÕES NO  CONJUNTO DE TESTE. FAZEMOS ISSO UTILIZANDO A FUNÇÃO predict() ASSOCIADA A INICIALIZAÇÃO DA FUNÇÃO GaussianNB(). ESSA FUNÇÃO DENOMINADA predict() RETORNA UMA MATRIZ DE PREVISÕES PARA CADA INSTÂNCIA DE DADOS NO CONJUNTO DE TESTES

print(preds) # AQUI IMPRIMIMOS NOSSAS PREVISÕES PARA TER UMA IDEIA DE O QUE O MODELO DETERMINOU

# AVALIANDO A PREVISÃO
print(accuracy_score(test_labels, preds)) # AQUI TRARÁ A PORCENTAGEM DE PRECISÃO DO CLASSIFICADOR (TAXA DE PRECISÃO DE CLASSIFICAÇÃO)
```
